```python
id_list = df['id'].tolist()
```

# **Fetching Genres and Taglines**

```python
keys = pd.DataFrame()
```

```python
# Iterate through id_list
for i in id_list:
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key=47ebbc23b50b94c3f6f40009c71efd23&&language=en-US'.format(i))

    # Check if the request was successful
    if response.status_code == 200:
        # Extract the genres and tagline from the JSON response
        result_json = response.json()
        genres = result_json.get('genres', None)
        tagline = result_json.get('tagline', None)
        # Append to the main DataFrame
        keys = keys.append({'id':i,'genres': genres, 'tagline': tagline}, ignore_index=True)
```

```python
keys
```

```python
def extract_genre_names(genre_list):
    if genre_list is not None:
        return ', '.join(genre['name'] for genre in genre_list)
    else:
        return ''
# Apply the function to the 'genres' column
keys['genre_names'] = keys['genres'].apply(extract_genre_names)
```

```python
keys
```

```python
keys = keys[['id','tagline','genre_names']]
```

```python
df = df[['id','title','overview']]
```

# **Merging Movie_details with Genres and Taglines**

```python
movies =  pd.merge(df, keys, on='id', how='left')
```

```python
movies.drop_duplicates(inplace=True)
movies.reset_index(drop=True, inplace=True)
```
